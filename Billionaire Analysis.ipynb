# World Wide Billionaire Analysis

```python
# importing necessary python libraires

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing dataset

df = pd.read_csv('Billionaire.csv')
print(df.head())
```

```python
# checking whether the dataset contains missing values

print(df.isnull().sum())
```

Here we can see that Age column has 79 missing values.

```python
# removing rows having missing values.

df = df.dropna()
```

```python
# The networth column in dataset contains $ sign at start and B at the end and converting it to float

df["NetWorth"] = df["NetWorth"].apply(str).str.strip("$")
df["NetWorth"] = df["NetWorth"].apply(str).str.strip("B")
df["NetWorth"] = df["NetWorth"].astype(float)

print(df.head())
```

### Top 10 Billionaires according to their Networth

```python
a = df.sort_values(by=["NetWorth"], ascending= False).head(10)
plt.figure(figsize=(20,10))
sns.histplot(x="Name", hue="NetWorth", data=a)
plt.show()
```

### Top 5 domain with most number of Billionires

```python
s = df["Source"].value_counts().head()
index = s.index
sources = s.values
custom_colors = ["blue", "green", 'pink', "yellow", "red"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors, autopct='%1.0f%%', pctdistance=0.7, labeldistance=1.1)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Domains with most no. of Billionaire", fontsize=15)
plt.show()
```

### Top 5 industries with most number of Billionaires

```python
s = df["Industry"].value_counts().head()
index = s.index
sources = s.values
custom_colors = ["grey", "lightgreen", 'pink', "yellow", "skyblue"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors, autopct='%1.0f%%', pctdistance=0.7, labeldistance=1.1)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries with most no. of Billionaire", fontsize=15)
plt.show()
```

### Top 5 Countries with most number of Billionaires 

```python
s = df["Country"].value_counts().head()
index = s.index
sources = s.values
custom_colors = ["red", "blue", 'brown', "green", "purple"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors, autopct='%1.0f%%', pctdistance=0.7, labeldistance=1.1)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries with most no. of Billionaire", fontsize=15)
plt.show()
```
